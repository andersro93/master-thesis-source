# Dynamic S results plots

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# Load the results
e_1 = pd.read_csv("./dynamic_s/1.csv", sep=";")
e_2 = pd.read_csv("./dynamic_s/2.csv", sep=";")
e_3 = pd.read_csv("./dynamic_s/3.csv", sep=";")
e_4 = pd.read_csv("./dynamic_s/4.csv", sep=";")
```

## Summary

```python
# Plot the adjustment of S during all of the training
## Lines
plt.plot(e_1.s, label="Experiment 1", color="blue")
plt.plot(e_2.s, label="Experiment 2", color="red")
plt.plot(e_3.s, label="Experiment 3", color="green")
plt.plot(e_4.s, label="Experiment 4", color="yellow")
plt.axhline(y=3.9, label="Optimal S", color="orange")

## Configuration
plt.title("Experiments")
plt.xlabel("Epoch")
plt.ylabel("S value");
plt.legend();

## Run the plot
plt.show()
```
