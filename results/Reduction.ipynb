# Reduction results plots

```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.float_format', lambda x: '%.10f' % x)

plt.style.use('seaborn-whitegrid')
```

```python
# Load the results
cpu_reduction = pd.read_csv("./reduction/reduce_cpu_results.csv", sep=";")
gpu_reduction = pd.read_csv("./reduction/reduce_gpu_results.csv", sep=";")
```

# CPU Reduction

```python
cpu_reduction.describe()
```

```python
cpu_reduction.mean().plot()
```

# GPU Reduction

```python
gpu_reduction.describe()
```

```python
gpu_reduction.mean().plot()
```

# Comparison

```python
## Lines
plt.plot(cpu_reduction.mean(), label="CPU", color="blue")
plt.plot(gpu_reduction.mean(), label="GPU", color="green")

## Configuration
plt.title("Experiments")
plt.xlabel("Dataset size expressed as power of 2")
plt.ylabel("Seconds used to calculate");
plt.legend();

## Axis
plt.xscale("linear")

plt.yticks([0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1.0, 2.0])
plt.yscale("log")

## Run the plot
plt.show()
```
